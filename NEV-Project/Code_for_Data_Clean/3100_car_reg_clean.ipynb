# Introduction

In this NoteBook we will clean the dataset of total number of registered vehicles per state in 2015 & 2020

# Set up

```python
import pandas as pd
```

# Read the data

```python
data = pd.read_csv('../Raw_Data/Total_number_of_registered_vehicles_per_state_2015&2020.csv',
                   thousands=',',
                   skiprows=[1], 
                   delimiter = ',',
                   index_col=[0]
                  )
data.set_axis(['2020', '2015'], axis=1, inplace=True)
data.head()
```

**Read population data to calculate per capita value**

```python
population = pd.read_csv('../Processed_Data/population.csv', delimiter = ',', index_col=0)
population.head()
```

```python
car_reg_2015_and_2020_per_capita = data/population[['2020','2015']]
car_reg_2015_and_2020_per_capita.head()
```

# Output Cleaned Data Version

```python
car_reg_2015_and_2020_per_capita.to_csv('../Processed_Data/car_reg_2015_and_2020_per_capita.csv')
data.to_csv('../Processed_Data/car_reg_2015_and_2020.csv')
```
