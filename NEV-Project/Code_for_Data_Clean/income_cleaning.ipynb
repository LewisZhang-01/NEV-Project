# Introduction

In this Notebook, we will clean the data of personal income of different states

## Set up

```python
import pandas as pd
```

## Read the Data

```python
#read the data file, skip the title and note 
data = pd.read_csv('../Raw_Data/income.csv', skiprows=[0,1,2,3,65,66,67,68,69], delimiter = ',')
```

```python
data.head()
```

## Data clean

```python
# we don't need column "GeoFips" , so clean this column
data = data.drop(data.columns[[0]], axis = 1)
```

```python
data.head()
```

there all some (NA) value in the dataset replace them with 0 value

```python
# some element has "(NA)" value, clean these as value 0
data = data.replace("(NA)",0)
```

```python
data
```

```python
data.iloc[2,0] = 'Alaska'
data.iloc[12,0] = 'Hawaii'
```

```python
#drop 'GeoName' column
col = list(data)
col.pop(0)
print(col)
```

### The data is quarterly related, we need it for each year

```python
#convert the element in DF to float
for i in col:
    data[i] = data[i].astype(float)
```

```python
#Sum for each 4 columns, create a column to store result, and drop those 4 columns
for i in range(1948,2022):
    data[i] = data.iloc[:,i-1947:i-1943].sum(axis=1)
    data = data.drop(data.columns[[1,2,3,4]],axis=1)
```

```python
data = data.drop([0,52,53,54,55,56,57,58,59])
data
```

```python
#save the cleaned data to processed file
path = '../Processed_Data/incomes.csv'
data.to_csv(path)
```
