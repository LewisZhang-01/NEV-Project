# Introduction

In this NoteBook we will clean the dataset of Number of gas stations.

## Set up

```python
import pandas as pd
```

## Read the data

```python
data = pd.read_csv('../Raw_Data/gas_station.csv', skiprows=[0,53], delimiter = ',')
```

```python
data.head()
```

```python
data.iloc[0,0]
```

### We need to change str to int

```python
a = list(data)
a.remove('State')
print(a)
```

```python
for year in a:
    data[year] = pd.to_numeric(data[year].astype(str).str.replace(',', ''))
```

```python
data.head()
```

```python
path = '../Processed_Data/gas_station.csv'
data.to_csv(path)
```
