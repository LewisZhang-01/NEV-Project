# Introduction

In this NoteBook we will clean the dataset of population of each state.

## Set up

```python
import pandas as pd
```

## Read the data

```python
data = pd.read_excel('../Raw_Data/population.xlsx',
                   thousands=',',
                   header=3,
                   #skiprows=lambda x: (x != 0) and not x % 2, 
                   index_col=0
                  )\
         #.iloc[2:]
```

```python
data.dropna(how='all', inplace=True)
data.head()
```

```python
data.index = data.index.str.replace('.', '',regex=True)
data.drop(['United States', 'Northeast','Midwest','South','West'], axis=0, inplace=True)
data.shape
```

```python
path = '../Processed_Data/population.csv'
data.to_csv(path)
```
