### Load the US shapefile

```python
US_gpd = gpd.read_file(USA_SHAPEFILE)[['NAME', 'geometry']]\
    .rename(columns={'NAME':'States'})\
    .set_index('States')

def get_Coords(df_Main):
    df_Main['coords'] = df_Main['geometry'].apply(lambda x: x.representative_point().coords[:])
    df_Main['coords'] = [coords[0] for coords in df_Main['coords']]
    return df_Main

US_gpd = get_Coords(US_gpd)
US_gpd.head()
```

## Draw New Energy Station Map
Add the new energy station data into the map dataframe as follows:

```python
# Change location by position dict
df_new_pos = pd.DataFrame.from_dict(new_position,orient='index')
df_old_pos = pd.DataFrame.from_dict(old_position,orient='index')
df_arrow_pos = pd.DataFrame.from_dict(arrow_position,orient='index')

# Construct tuple for coords with x and y axis position
df_new_pos['coords']=df_new_pos[[0,1]].apply(tuple,axis=1)
df_old_pos['old_coords']=df_old_pos[[0,1]].apply(tuple,axis=1)
df_arrow_pos['arrow_coords']=df_arrow_pos[[0,1]].apply(tuple,axis=1)

# Drop the useless columns
df_new_pos = df_new_pos.drop(columns=[0,1])
df_old_pos = df_old_pos.drop(columns=[0,1])
df_arrow_pos = df_arrow_pos.drop(columns=[0,1])

# Update the original table
US_gpd.update(df_new_pos)

# Merge new & old position in one df
df_overlap_pos = pd.concat([df_new_pos['coords'].rename('new'), \
                            df_old_pos['old_coords'].rename('old'),\
                            df_arrow_pos['arrow_coords'].rename('arrow')],\
                           axis = 1)
```

```python
US_NEV_per_1k = pd.concat([(df_NEV[df_NEV['Year']==2012]['Total']/(df_pop['2012']/1000)).rename('per_1k_people'),
                            US_gpd], axis = 1)
US_NEV_per_1k['State'] = US_NEV_per_1k.index
US_NEV_per_1k.head()
```

```python
def draw_US_map_with_label(max_,min_,US_Station,title,bar_title,c_ax,col_name):
    fig = plt.figure(figsize=[16,12],dpi=100)
    ax = fig.add_subplot(111)
    
    # Convert into GeoDF fromat
    US_Station = GeoDataFrame(US_Station)
    
    # Split into 3 part to display the map
    df_Alaska = US_Station[US_Station.index == 'Alaska']
    df_Hawaii = US_Station[US_Station.index == 'Hawaii']
    df_Main = US_Station.drop(index=('Alaska'))
    df_Main = df_Main.drop(index=('Hawaii'))
    
    # Setup the colour map and normalise it.
    norm = matplotlib.colors.Normalize(vmin=min_, vmax=max_)
    cmap = matplotlib.cm.get_cmap('Spectral_r')

    # Draw US main map
    df_Main.plot(column=col_name, cmap=cmap, norm=norm, ax=ax, lw=.25, edgecolor='darkslategrey',\
                 missing_kwds={"color": "lightgrey","edgecolor": "red","hatch": "///","label": "Missing values"})

    # Draw state names
    for idx, row in df_Main.iterrows():
       plt.annotate(text=row['State'], xy=row['coords'], horizontalalignment='center', color='black')
    # Draw arrow for overlapped name
    for idx, row in df_overlap_pos.iterrows():
        ax.annotate("", xy=row['arrow'], xytext=row['old'],arrowprops=dict(arrowstyle="->"))
    
    # Tighten up the view
    plt.xlim([-140, -55])
    plt.ylim([13, 55])

    # Remove unnecessary tick marks.
    remove_tick_mark(ax)

    # Set title label.
    ax.set_title(title)

    # Add the colour bar and label it.
    cax = fig.add_axes(c_ax)
    cb = matplotlib.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm, spacing='proportional', orientation='horizontal')
    cb.set_label(bar_title, fontsize=14)

    # Draw Alaska Map
    ax = fig.add_subplot(4,5,6)
    df_Alaska.plot(column=col_name, cmap=cmap, norm=norm, ax=ax, lw=.25, edgecolor='darkslategrey',\
                   missing_kwds={"color": "lightgrey","edgecolor": "red",\
                                 "hatch": "///","label": "Missing values"}\
                  )
    # Tighten up the view
    plt.xlim([-180, -128])
    plt.ylim([50,75])
    # Show Alaska name
    plt.annotate(text='Alaska', xy=df_Alaska['coords']['Alaska'], horizontalalignment='center', color='black')
    # Remove unnecessary tick marks.
    remove_tick_mark(ax)
    
    # Draw Hawaii Map
    ax = fig.add_subplot(4,5,11)
    df_Hawaii.plot(column=col_name, cmap=cmap, norm=norm, ax=ax, lw=.25, edgecolor='darkslategrey',\
                   missing_kwds={"color": "lightgrey","edgecolor": "red",\
                                 "hatch": "///","label": "Missing values"}\
                  )
    # Show Hawaii name
    plt.annotate(text='Hawaii', xy=df_Hawaii['coords']['Hawaii'], horizontalalignment='center', color='black');
    # Remove unnecessary tick marks.
    remove_tick_mark(ax)
    
    return fig

# Remove unnecessary tick marks.
def remove_tick_mark(ax):
    ax.tick_params(axis=u'both', which=u'both',length=0)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
        

fig = draw_US_map_with_label(max_ = US_NEV_per_1k['per_1k_people'].sort_values(ascending=False).head(10).mean(),
                             min_ = US_NEV_per_1k['per_1k_people'].min(),
                             US_Station = US_NEV_per_1k,
                             title = 'Distribution of new energy stations per thousand people in the US (2012)',
                             bar_title = 'New Energy station number per 1k people',
                             # Set colour bar position.
                             c_ax = [.3, .32, .4, .01],
                             col_name = 'per_1k_people')

# Save as png image
fig.savefig('./Outputs/NEV_Station_per_1k_people.png', format='png')
```

# 2007-2012 Station Trands

```python
df_NEV_07_12 = df_NEV[(df_NEV['Year']>=2007)&(df_NEV['Year']<=2012)][['Total','Year']]
df_NEV_07_12 = df_NEV_07_12.pivot(columns ='Year')['Total']
df_NEV_07_12.head()
```

```python
df_Gas_07_12 = df_Gas[[str(int_) for int_ in list(range(2007, 2013))]]
df_Gas_07_12.head()
```

```python
df_pop_07_12 = df_pop[[str(int_) for int_ in list(range(2007, 2013))]]
df_pop_07_12.head()
```

```python
# df_NEV_07_12 = df_NEV_07_12/df_pop_07_12
# df_Gas_07_12 = df_Gas_07_12/df_pop_07_12
```

```python
nev_diff = pd.DataFrame(df_NEV_07_12.T.diff().sum())
gas_diff = pd.DataFrame(df_Gas_07_12.T.diff().sum())

all_diff = pd.concat([nev_diff[0].rename('nev'),gas_diff[0].rename('gas'),US_gpd],axis=1)
all_diff['State'] = all_diff.index
all_diff.head()
```

```python
both_neg = all_diff[(all_diff['nev'].values < 0) & (all_diff['gas'].values < 0)]
nev_pos_gas_neg = all_diff[(all_diff['nev'].values > 0) & (all_diff['gas'].values < 0)]
nev_neg_gas_pos = all_diff[(all_diff['nev'].values < 0) & (all_diff['gas'].values > 0)]
both_pos = all_diff[(all_diff['nev'].values > 0) & (all_diff['gas'].values > 0)]

def check_group(state_name,both_neg,both_pos,nev_pos_gas_neg,nev_neg_gas_pos):
    if state_name in both_neg.index:
        return 'both_neg'
    elif state_name in both_pos.index:
        return 'both_pos'
    elif state_name in nev_pos_gas_neg.index:
        return 'nev_pos_gas_neg'
    elif state_name in nev_neg_gas_pos.index:
        return 'nev_neg_gas_pos'
```

```python
def draw_US_map_with_label_for_delta(max_,min_,US_Station,title,bar_title,c_ax,column_name,both_neg,nev_pos_gas_neg,nev_neg_gas_pos,both_pos):
    fig = plt.figure(figsize=[16,12],dpi=100)
    ax = fig.add_subplot(111)
    
    # Convert into GeoDF fromat
    US_Station = GeoDataFrame(US_Station)
    
    # Split into 3 part to display the map
    df_Alaska = US_Station[US_Station.index == 'Alaska']
    df_Hawaii = US_Station[US_Station.index == 'Hawaii']
    df_Main = US_Station.drop(index=('Alaska'))
    df_Main = df_Main.drop(index=('Hawaii'))
    
    df_Alaska_trend = check_group('Alaska',both_neg,both_pos,nev_pos_gas_neg,nev_neg_gas_pos)
    df_Hawaii_trend = check_group('Hawaii',both_neg,both_pos,nev_pos_gas_neg,nev_neg_gas_pos)
    
    # Setup the colour map and normalise it.
    norm = matplotlib.colors.Normalize(vmin=min_, vmax=max_)
    cmap = matplotlib.cm.get_cmap('Spectral_r')

    # Draw US main map
    df_Main.plot(column=column_name, cmap=cmap, norm=norm, ax=ax, lw=.25, edgecolor='darkslategrey',\
                 missing_kwds={"color": "lightgrey","edgecolor": "red","hatch": "///","label": "Missing values"})
    
    
    c1=both_neg.plot(ax=ax,color='none',alpha=0.5,hatch= "x")
    c2=both_pos.plot(ax=ax,color='none', edgecolor='grey',alpha=0.5,hatch= "*")
    c3=nev_pos_gas_neg.plot(ax=ax,color='none', edgecolor='grey',alpha=0.5,hatch= "+")
    c4=nev_neg_gas_pos.plot(ax=ax,color='none', edgecolor='grey',alpha=0.5,hatch= ".")
    
    #ax.legend(handles = [c1,c2,c3,c4],loc=2)
    
    # Draw state names
    for idx, row in df_Main.iterrows():
       plt.annotate(text=row['State'], xy=row['coords'], horizontalalignment='center', color='black')
    # Draw arrow for overlapped name
    for idx, row in df_overlap_pos.iterrows():
        ax.annotate("", xy=row['arrow'], xytext=row['old'],arrowprops=dict(arrowstyle="->"))
    
    # Tighten up the view
    plt.xlim([-140, -55])
    plt.ylim([13, 55])

    # Remove unnecessary tick marks.
    remove_tick_mark(ax)

    # Set title label.
    ax.set_title(title)

    # Add the colour bar and label it.
    cax = fig.add_axes(c_ax)
    cb = matplotlib.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm, spacing='proportional', orientation='horizontal')
    cb.set_label(bar_title, fontsize=14)

    # Draw Alaska Map
    ax = fig.add_subplot(4,5,6)
    df_Alaska.plot(column=column_name, cmap=cmap, norm=norm, ax=ax, lw=.25, edgecolor='darkslategrey',\
                   missing_kwds={"color": "lightgrey","edgecolor": "red",\
                                 "hatch": "///","label": "Missing values"}\
                  )
    if df_Alaska_trend == 'both_neg':
        df_Alaska.plot(ax=ax,color='none',alpha=0.5,hatch= "x")
    elif 'both_pos':
        df_Alaska.plot(ax=ax,color='none',alpha=0.5,hatch= "*")
    elif 'nev_pos_gas_neg':
        df_Alaska.plot(ax=ax,color='none',alpha=0.5,hatch= "+")
    elif 'nev_neg_gas_pos':
        df_Alaska.plot(ax=ax,color='none',alpha=0.5,hatch= ".")
        
    # Tighten up the view
    plt.xlim([-180, -128])
    plt.ylim([50,75])
    # Show Alaska name
    plt.annotate(text='Alaska', xy=df_Alaska['coords']['Alaska'], horizontalalignment='center', color='black')
    # Remove unnecessary tick marks.
    remove_tick_mark(ax)
    
    # Draw Hawaii Map
    ax = fig.add_subplot(4,5,11)
    df_Hawaii.plot(column=column_name, cmap=cmap, norm=norm, ax=ax, lw=.25, edgecolor='darkslategrey',\
                   missing_kwds={"color": "lightgrey","edgecolor": "red",\
                                 "hatch": "///","label": "Missing values"}\
                  )
    if df_Hawaii_trend == 'both_neg':
        df_Hawaii.plot(ax=ax,color='none',alpha=0.5,hatch= "x")
    elif 'both_pos':
        df_Hawaii.plot(ax=ax,color='none',alpha=0.5,hatch= "*")
    elif 'nev_pos_gas_neg':
        df_Hawaii.plot(ax=ax,color='none',alpha=0.5,hatch= "+")
    elif 'nev_neg_gas_pos':
        df_Hawaii.plot(ax=ax,color='none',alpha=0.5,hatch= ".")
        
    # Show Hawaii name
    plt.annotate(text='Hawaii', xy=df_Hawaii['coords']['Hawaii'], horizontalalignment='center', color='black');
    # Remove unnecessary tick marks.
    remove_tick_mark(ax)
    
    return fig

# Remove unnecessary tick marks.
def remove_tick_mark(ax):
    ax.tick_params(axis=u'both', which=u'both',length=0)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
        
# Setting for adjust color range
fig = draw_US_map_with_label_for_delta(max_ = all_diff['gas'].sort_values(ascending=False).head(1),
                                       min_ = all_diff['gas'].sort_values(ascending=True).head(1),
                                       US_Station = all_diff,
                                       title = 'Vehicle Station Development Delta Value Among 2007-2012 in US',
                                       bar_title = 'Gas Station Development Delta',
                                       # Set colour bar position.
                                       c_ax = [.3, .32, .4, .01],
                                       column_name = 'gas',
                                       both_neg = both_neg,
                                       nev_pos_gas_neg = nev_pos_gas_neg,
                                       nev_neg_gas_pos = nev_neg_gas_pos,
                                       both_pos = both_pos)
# Save as png image
fig.savefig('./Outputs/Vehicle_Station_Development_Delta_Value_Among_2007_2012.png', format='png')
```
