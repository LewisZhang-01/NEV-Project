# Load and aggregate Rating Data

```python
import json
from glob import glob
import pandas as pd
import numpy as np
import itertools

from loguru import logger
from IPython.display import display, clear_output
```

Path to rating dataset

```python
complaints_dataset = "../raw/Ratings"
```

Aggregate all json file in this directory.

```python
files = glob('../raw/Ratings/*.json')
len(files)
```

```python
data = []

for i, file in enumerate(files):
    
    clear_output()
    logger.info((i, file))
    
    with open(file, 'r') as f: data.append(json.load(f))
        
df = pd.DataFrame(itertools.chain.from_iterable(data))
df.shape
```

```python
rating_df_clean = df.copy()
rating_df_clean.describe()
```
