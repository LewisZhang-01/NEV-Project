```python
import pandas as pd
import numpy as np
import csv
from datetime import datetime
```

```python
#read the file
df = pd.read_csv('laws_and_incentives (Apr 7 2022).csv')
df.head()
```

```python
#show the columns and choose what we need (State,Type)
list(df)
```

```python
df=df.drop(['Law Id',

 'Title',
 'Text',
 'Enacted Date',
 'Amended Date',
 'Recent?',
 'Sequence Number',

 'Agency',
 'Significant Update Date',
 'Expired Date',
 'Archived Date',
 'Repealed Date',
 'Topic',
 'Technology Categories',
 'Incentive Categories',
 'Regulation Categories',
 'User Categories',
 'References'],axis=1)
```

```python
list=[]
for index, row in df.iterrows():
    if df.iloc[index,0] == 'US':
        list.append(index)
df = df.drop(list,axis=0)
```

```python
State = np.sort(['Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Delaware','Florida','Georgia','Hawaii',
             'Idaho','Illinois','Indiana','Iowa','Kansa','Kentucky','Louisiana','Maine','Maryland','Massachusetts','Michigan',
             'Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico','New York',
             'North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island','South Carolina','South Dakota','Tennessee','Texas',
             'Utah','Vermont','Virginia','Washington','West Virginia','Wisconsin','Wyoming'])
Postal_Abbr = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH',
    'OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']
```

```python
for index in range(50):
    df = df.replace(Postal_Abbr[index], State[index])
```

```python
df
```

```python
Type = np.sort(df['Type'].unique())
Type
```

```python
new_df = pd.DataFrame(columns=Type)
new_df.insert(loc=0,column='State',value=State)
new_df = new_df.set_index('State')
```

```python
new_df.head()
```

```python
for state in State:
    for tp in Type:
        count=0
        for index, row in df.iterrows():
            if df['State'][index] == state and df['Type'][index] == tp:
                count += 1
        new_df.at[state,tp] = count
new_df['Total'] = new_df.iloc[:,0:2].sum(axis=1)
```

```python
new_df.head()
```

```python
# Now we have cleanned data, save it into processed file
path = '../../processed/policy_state.csv'
new_df.to_csv(path)
```
